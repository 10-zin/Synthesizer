```python
def get_info(file_name):
    with open(file_name, 'r') as file:
        lines = file.readlines()
    info = []
    for i, line in enumerate(lines):
        if i==0:
            continue
#         print(line)
        info.append([float(i) for i in str(line).strip('b\n').split(',')])
        
    ppl = []
    acc = []
    loss = []
    for i in range(len(info)):
        ppl.append(info[i][2])
        acc.append(info[i][3])
        loss.append(info[i][1])
        
    return loss, ppl, acc

file_name = 'dense_1.train.log'
train_loss, train_ppl, train_acc = get_info(file_name)
file_name = 'dense_1.valid.log'
val_loss, val_ppl, val_acc = get_info(file_name)
file_name = 'random_1.train.log'
train_loss_rndm, train_ppl_rndm, train_acc_rndm = get_info(file_name)
file_name = 'random_1.valid.log'
val_loss_rndm, val_ppl_rndm, val_acc_rndm = get_info(file_name)
file_name = 'vanilla_1.train.log'
train_loss_vanilla, train_ppl_vanilla, train_acc_vanilla = get_info(file_name)
file_name = 'vanilla_1.valid.log'
val_loss_vanilla, val_ppl_vanilla, val_acc_vanilla = get_info(file_name)
```

```python
import matplotlib.pyplot as plt
def plot(y, y_2, y_3, y_label):
    plt.plot([i for i in range(15)], y, label = 'dense')
    plt.plot([i for i in range(15)], y_2, label = 'vanilla')
    plt.plot([i for i in range(15)], y_3, label = 'random')
    plt.ylabel(y_label)
    plt.xlabel('epochs')
    plt.legend()
    plt.show()
```

```python
plot(val_ppl[10:], val_ppl_vanilla[10:], val_ppl_rndm[10:], 'perplexity')
```

```python
plot(val_acc[10:], val_acc_vanilla[10:], val_acc_rndm[10:], 'accuracy')
```

```python
plot(val_loss[10:], val_loss_vanilla[10:], val_loss_rndm[10:], 'loss')
```
